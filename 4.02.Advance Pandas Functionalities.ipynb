# Pandas - Data Frame Operations

```python
import numpy as np
import pandas as pd
```

```python
# create a simple dataframe

data = pd.DataFrame({
    'age' :     [ 10, 22, 13, 21, 12, 11, 17],
    'section' : [ 'A', 'B', 'C', 'B', 'B', 'A', 'A'],
    'city' :    [ 'Gurgaon', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'gender' :  [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'favourite_color' : [ 'red', 'blue', 'yellow', 'blue', 'black', 'green', 'red']
})

# view the data
data
```

## loc and iloc

- **loc** gets columns (or rows) with particular **labels/names** from the index.
- **iloc** gets columns (or rows) at particular **positions/index** in the index (**so it only takes integers**).

```python
data.iloc[1:3,2]
```

### Select rows with indices using iloc

```python
# select rows with indexes

data.iloc[[0,2]]

# 1st the row and the 3rd row
```

### Select rows with particular indices and particular columns == Slicing

```python
display(data)
data.iloc[[0,2],[1,3]]
```

### Select a range of rows and columns using iloc

```python
display (data)
data.iloc[1:3,2:4]
```

```python
data.iloc[:, :4]
```

### Find all the rows based on any condition in a column

We can solve types of queries with a simple line of code using pandas.DataFrame.loc[]. We just need to pass the condition within the loc statement.

```python
display(data)
# data.age > 15
data.loc[data.age > 15]
```

### Find all the rows with more than one condition in columns

```python
display(data)
data.loc[(data.age >= 12) & (data.gender == "M")]
```

### Select only required columns with a condition

```python
data.loc[(data.age >= 12), ['city', 'gender']]
```

### Update the values of a particular column on selected rows

We can do this by running a for loop as well but if our dataset is big in size, then it would take forever to complete the task. Using loc in Pandas, we can do this within seconds, even on bigger datasets!

We just need to specify the condition followed by the target column and then assign the value with which we want to update

```python
display(data)
```

```python
display(data.loc[(data.age >= 15),["age","section"]])
data.loc[(data.age >= 15),["section"]] = "UpdateS"
display(data.loc[(data.age >= 15),["age","section"]])
```

```python
display(data.loc[(data.age >= 20),['age',"favourite_color","city"]])
data.loc[(data.age >= 20),["favourite_color","city"]] = ['BLACK', "Dubai"]
display(data.loc[(data.age >= 20),['age',"favourite_color","city"]])
```

## Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
data
```

```python
df = pd.DataFrame(data)
df
```

**Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:**

```python
# We want to look at average sales companywise
df.groupby('Company')
```

```python
df.groupby('Company').mean()
```

```python
df.groupby('Company').count()
```

## Performing Merge, Join and Concatenate DataFrames in Pandas

**Merging is the Pandas operation that performs database joins on objects**

```python
dfExample1 = pd.DataFrame([[4,'QA'],[1,'HR'],[3,'Dev'],[2,'Ops']] , columns=['eid','dept'])
display (dfExample1)

dfExample2 = pd.DataFrame([[1,'Prashant'],[2,'Gokul'],[3,'Guna']] , columns=['eid','ename'])
display (dfExample2)
```

```python
# Use-case. I want to merge two dfs based on eid
# We will use merge

resultDF = pd.merge(dfExample2, dfExample1)
display (resultDF)

# The merge worked in this case because both dataframe have the common column eid
# that too with the same name
```

```python
resultDF = pd.merge(dfExample1,dfExample2)
display (resultDF)
```

**Left Join : Returns all rows from the left table, even if there are no matches in the right table**

```python
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'], 'ICC_rank':[2,3,7,8,4],      
                 'World_champions_Year':[2011,2015,1979,1992,1996], 'Points':[874,787,753,673,855]} 

chokers={'Team':['South Africa','New Zealand','Zimbabwe'], 'ICC_rank':[1,5,9],'Points':[895,764,656]} 

df1=pd.DataFrame(world_champions) 
df2=pd.DataFrame(chokers) 

display(df1) 

display(df2) 
```

```python
pd.merge(df1, df2, on = "Team")
# empty df = nothig common in Teams
```

```python
pd.merge(df1, df2, on = "Team", how = 'left')
# suffice x - col is from left table; suffice y - col is from right table 
```

**Right Join : Preserves the unmatched rows from the second (right) table, joining them with a NULL in the shape of the first (left) table**

```python
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'], 'ICC_rank':[2,3,7,8,4],         
                 'World_champions_Year':[2011,2015,1979,1992,1996], 'Points':[874,787,753,673,855]} 
chokers={'Team':['South Africa','New Zealand','Pakistan'],'ICC_rank':[1,5,9],'Points':[895,764,656]} 

df1=pd.DataFrame(world_champions) 
df2=pd.DataFrame(chokers) 

display(df1) 
display(df2) 
```

```python
pd.merge(df1, df2, on='Team', how='right')
# suffice x - col is from left table; suffice y - col is from right table 
```

**Inner Join : Selects all rows from both participating tables if there is a match between the columns**

```python
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'], 'ICC_rank':[2,3,7,8,4],         
                 'World_champions_Year':[2011,2015,1979,1992,1996], 'Points':[874,787,753,673,855]} 
chokers={'Team':['South Africa','New Zealand','Pakistan'],'ICC_rank':[1,5,9],'Points':[895,764,656]} 

df1=pd.DataFrame(world_champions) 
df2=pd.DataFrame(chokers) 

display(df1) 
display(df2) 
```

```python
pd.merge(df1,df2, on='Team', how='inner')
```

**Full Outer Join : Returns all records when there is a match in either left (table1) or right (table2) table records**

```python
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'], 'ICC_rank':[2,3,7,8,4],         
                 'World_champions_Year':[2011,2015,1979,1992,1996], 'Points':[874,787,753,673,855]} 
chokers={'Team':['South Africa','New Zealand','Zimbabwe'],'ICC_rank':[1,5,9],'Points':[895,764,656]} 

df1=pd.DataFrame(world_champions) 
df2=pd.DataFrame(chokers)

display(df1) 
display(df2) 
```

```python
display(pd.merge(df1,df2, on='Team', how = 'outer'))
```

### Concat Operation

```python
empExample = pd.DataFrame([[1,'Prashant'],[2,'Gokul'],[3,'Guna']] , columns=['empid','ename'])

empExample2 = pd.DataFrame([[4,'Nik'],[5,'Ashish'],[6,'Asha']] , columns=['empid','ename'])

display (empExample)
display (empExample2)
```

```python
# Rowwise Concatenation --- Ensure the column names are same in all DFs.

resultEmp = pd.concat([empExample,empExample2],ignore_index=True)
resultEmp
```

```python
resultEmp = pd.concat([empExample,empExample2],ignore_index=True, axis = 1)
resultEmp
```
