```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_auto=pd.read_csv('C:\\Users\\Anutthamaa\\PracticePrograms\\Data sets\\OSL Datasets\\auto_data_mpg.csv')
df_auto.head()
```

```python
#user defined function to replace 1 as USA 2 as EUROPE and 3 as ASIA in origin column
def origin(num):
    if (num==1):
        return 'USA'
    elif(num==2):
        return 'EUROPE'
    else:
        return 'ASIA'
df_auto['origin']=df_auto['origin'].apply(origin)
    
```
