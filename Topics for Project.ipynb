```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
bs = load_boston()
bs.keys()
```

```python
print(bs.DESCR)
```

```python
bs.data.shape
```

```python
boston = pd.DataFrame(bs.data, columns = bs.feature_names)
boston['MEDV'] = bs.target

boston.head()
```

# Correlation

## Heat Map : Correlation Matrix

```python
boston.info()
```

```python
boston.corr()
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(boston.corr(),annot = True, annot_kws= {'size' : 12},
           cmap = "PiYG", fmt = '.2f', linewidths = 2, linecolor = 'white');
```

 **with threshold 0.7 or -0.7**

```python
boston.columns
```

```python
sns.pairplot(boston, vars = ['CRIM', 'NOX', 'RM', 'AGE', 'DIS', 'LSTAT', 'MEDV'], diag_kind="kde");
```

```python
boston[['CRIM', 'NOX', 'RM', 'AGE', 'DIS', 'LSTAT', 'MEDV']].skew()
```

- A skewness value greater than 1 or less than -1 indicates a highly skewed distribution. 
- A value between 0.5 and 1 or -0.5 and -1 is moderately skewed. 
- A value between -0.5 and 0.5 indicates that the distribution is fairly symmetrical.
