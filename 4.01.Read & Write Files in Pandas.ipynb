### How to drop/delete a column?

```python
data_clean = pd.read_csv("winequality_clean.csv")
display (data_clean.head()) 
```

```python
data_clean.drop('Unnamed: 0', axis = 1)

# does not update the dataframe == the cols will still be in the orignal df
```

with drop 
- axis = 1 refers to cols
- axis = 0 refers to rows

```python
data_clean.head(2)
```

```python
data_clean.drop('Unnamed: 0', axis = 1, inplace = True)
# inplace = True :: the changes will be reflected in the DF
```

```python
data_clean.head(2)
```

```python
data_clean.drop(columns = "ID", inplace = True)
data_clean.head(2)
```

```python
data_clean.drop(columns = ["free sulfur dioxide", "total sulfur dioxide"], inplace = True)
data_clean.head(2)
```

```python
data_clean.drop(["density", "pH"], axis = 1, inplace = True)
data_clean.head(2)
```

### Delete the rows

```python
data_clean.drop(0, axis = 0, inplace = True)
data_clean.head()
```

```python
data_clean.index[0:50]
```

```python
data_clean.drop(data_clean.index[0:50], axis = 0, inplace = True)
data_clean.head()
```
