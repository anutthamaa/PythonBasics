```python
import pandas as pd
import numpy as np
df1=pd.read_csv('C:\\Users\\Anutthamaa\\PracticePrograms\\Data sets\\OSL Datasets\\class-grades.csv')
#df1.to_csv-writes to csv file
```

```python
#to find null values in dataset output is column values
#returns true if column has null values and false for vice versa
df1.isna().any()
```

```python
#replacing the missing values with variable's mean
from sklearn.impute import SimpleImputer
mean_imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
mean_imputer=mean_imputer.fit(df1)
imputed_df=mean_imputer.transform(df1.values)
df1=pd.DataFrame(data=imputed_df,columns=['Prefix','Assignment','Tutorial','Midterm','TakeHome','Final'])
df1.head(23)
#print(df1[19:50])
```

```python
#replacing the missing values with variable's median
from sklearn.impute import SimpleImputer
median_imputer=SimpleImputer(missing_values=np.nan,strategy='median')
median_imputer=median_imputer.fit(df1)
imputed_df=median_imputer.transform(df1.values)
df1=pd.DataFrame(data=imputed_df,columns=['Prefix','Assignment','Tutorial','Midterm','TakeHome','Final'])
df1.head(9)
```

```python
#to detect outliers
import seaborn as sns
sns.boxplot(x=df1['Assignment'])#plot showing values<60 are outliers
```

```python
#create a filter based on box plot obtained and apply filter to the dataframe
filter=df1['Assignment'].values>60
df1_outlier_rem=df1[filter]
df1_outlier_rem
```
