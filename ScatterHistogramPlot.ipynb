```python
#import libraries
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#fetch california housing data
california_ds=fetch_california_housing()
print(california_ds['DESCR'])
```

```python
#define the x-axis and y-axis
x_axis=california_ds.data
y_axis=california_ds.target
```

```python
#histogram plot
style.use('ggplot')
plt.figure(figsize=(10,10))
plt.hist(y_axis,bins=50)
plt.xlabel('Median House values')
plt.ylabel('Prices in USD')
plt.show()
```

```python
#scatter plot
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(california_ds.data[:,7],california_ds.target)
plt.xlabel('Median House values')
plt.ylabel('Prices in USD')
plt.show()
```

```python
#xy plot 
#website traffic data
#no of website users
website_users=[123,456,789,112,345,678,901,234]
#time distribution hourly
#123 customers ar 2am, 456 customers at 4am and so on
time_hrs=[2,4,6,7,8,10,12,17]


style.use('ggplot')
#time_hrs on x axis and website users on Y axis
plt.plot(time_hrs,website_users,linestyle='--',linewidth=2.5,color='b',label='website traffic')
#plt.axis([5,7.5,500,800]) #set or define the range on x-axis and y-axis
#annotates the graph showing the max value
plt.annotate('Max',ha='center',va='bottom',xytext=(8,900),xy=(12,900),arrowprops={'facecolor':'green'})
plt.title('Website Traffic')
plt.xlabel('Time in Hrs')
plt.ylabel("No.of users")
plt.legend()
plt.show()
```

```python
#website traffic data
#no of website users on monday
website_monday_users=[123,456,789,112,345,678,901,234]
#no of website users on tuesday
website_tuesday_users=[111,222,333,444,555,666,777,888]
#no of website users on wednesday
website_wednesday_users=[100,200,300,400,500,600,700,800]
#time distribution hourly
#123 customers ar 2am, 456 customers at 4am and so on
time_hrs=[2,4,6,7,8,10,12,17]


style.use('ggplot')
#time_hrs on x axis and website users on Y axis
plt.plot(time_hrs,website_monday_users,linewidth=1.5,color='g',label='monday')
plt.plot(time_hrs,website_tuesday_users,linewidth=2.5,color='b',label='tuesday')
plt.plot(time_hrs,website_wednesday_users,linewidth=2.0,color='r',label='wednesday')
plt.axis([0,17.5,100,1000]) #set or define the range on x-axis and y-axis
#annotates the graph showing the max value
plt.annotate('Max',ha='center',va='bottom',xytext=(8,900),xy=(12,900),arrowprops={'facecolor':'green'})
plt.title('Website Traffic')
plt.xlabel('Time in Hrs')
plt.ylabel("No.of users")
plt.legend()
plt.show()
```
