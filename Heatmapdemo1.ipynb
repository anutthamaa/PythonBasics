```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flight_data=sns.load_dataset('flights')
flight_data.head()
```

```python
final_flight_data=flight_data.pivot('month','year','passengers')
final_flight_data
```

```python
#plot heat map 
sns.heatmap(final_flight_data)
```

```python
#pie chart 
jobdata=['100','90','40','20','10','12']
labels='IT','Finance','Marketing','Admin','HR','Operations'
explode=(0.05,0,0,0,0,0)
plt.pie(jobdata,labels,explode=explode)
plt.show()
```
